```python
# from the above graph we can fillup nan values with more_count_worker in each category
# to avoid overfitting we may use any one of three in case of Telecommunications
def fill_job_title(columns):
    job_title = columns[0]
    job_industry_category = columns[1]
    
    if pd.isnull(job_title):
        
        if job_industry_category == 'Manufacturing' :
            return 'Chemical Engineer'
        
        elif job_industry_category == 'Financial Services' :
            return 'Cost Accountant'
        
        elif job_industry_category == 'Health' :
            return 'Social Worker'
        
        elif job_industry_category == 'Retail' :
            return 'Sales Representative'
        
        elif job_industry_category == 'Property' :
            return 'Account Coordinator'
        
        elif job_industry_category == 'IT' :
            return 'Business Systems Development Analyst'
        
        elif job_industry_category == 'Entertainment' :
            return 'Assistant Media Planner'
        
        elif job_industry_category == 'Argiculture' :
            return 'Executive Secretary'
        
        elif job_industry_category == 'Telecommunications' :
            return 'Desktop Support Technician'
        
        else :
            return 'Nan'
        
    else:
        return job_title
```

```python
df1['job_title'] = df1[['job_title','job_industry_category']].apply(fill_job_title,axis=1)
```

```python
def plot_correlation_map( dff ):
    corr = dff.corr()
    _ , ax = plt.subplots( figsize =( 12 , 10 ) )
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
    _ = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
    )
```
